## 4. Summarizing Data

**Research Question**: How does the relationship between time spent gaming daily and academic performance change based on different values of sleep quality?

**Methodology**: To assess this research question, we will compare gamers in the dataset with a sleep quality that is very poor, fair, and good. Within each sleep quality group, we will look at how the distribution of hours spent gaming daily compares across academic performance levels. Then, we will compare the graphs from each level of sleep quality to see if there is a shift in daily gaming hours across all academic performance levels as sleep quality increases.

### Distribution of Hours Spent Gaming Daily by Academic Performance Level *(Very Poor Sleep)*
First, let's assess the correlation between time spent gaming daily and academic performance for individuals in the dataset who have ***very poor*** sleep quality.

```python
df_very_poor = df[df["sleep_quality"] == "Very Poor"]
sns.boxplot(x = df_very_poor["daily_gaming_hours"], y = df_very_poor["academic_work_performance"], order = ["Excellent", "Good", "Average", "Below Average", "Poor", "Failing"])
plt.title("Daily Gaming Duration vs. Academic Performance for Individuals with Very Poor Sleep Quality")
plt.xlim([0, 15])
summary1 = pd.DataFrame(df_very_poor["daily_gaming_hours"].describe()).reset_index()
summary1.columns = ["Statistic Type", "Very Poor Quality Sleep"]
summary1
```

### Distribution of Hours Spent Gaming Daily by Academic Performance Level *(Fair Sleep)*

Next, let's assess the correlation between time spent gaming daily and academic performance for individuals in the dataset who have ***fair*** sleep quality.

```python
df_fair = df[df["sleep_quality"] == "Fair"]
sns.boxplot(x = df_fair["daily_gaming_hours"], y = df_fair["academic_work_performance"], order = ["Excellent", "Good", "Average", "Below Average", "Poor", "Failing"])
plt.title("Daily Gaming Duration vs. Academic Performance for Individuals with Fair Sleep Quality")
plt.xlim([0, 15])
summary2 = pd.DataFrame(df_fair["daily_gaming_hours"].describe()).reset_index()
summary2.columns = ["Statistic Type", "Fair Quality Sleep"]
summary2
```

### Distribution of Hours Spent Gaming Daily by Academic Performance Level *(Good Sleep)*

Lastly, let's assess the correlation between time spent gaming daily and academic performance for individuals in the dataset who have ***good*** sleep quality.

```python
df_good = df[df["sleep_quality"] == "Good"]
sns.boxplot(x = df_good["daily_gaming_hours"], y = df_good["academic_work_performance"], order = ["Excellent", "Good", "Average", "Below Average", "Poor", "Failing"])
plt.title("Daily Gaming Duration vs. Academic Performance for Individuals with Good Sleep Quality")
plt.xlim([0, 15])
summary3 = pd.DataFrame(df_good["daily_gaming_hours"].describe()).reset_index()
summary3.columns = ["Statistic Type", "Good Quality Sleep"]
summary3
```

### Hours Spent Gaming Daily by Sleep Quality
*(Not broken down into levels of academic work performance, as shown in graphs above.)*

```python
summary1.merge(summary2, on = "Statistic Type").merge(summary3, on = "Statistic Type")
```

### Findings

Since each visualization above is standardized on a x-axis range of 0 to 15 hours of gaming daily, we can easily compare the relationships seen between daily hours gamed and academic performance under each level of sleep quality (very poor, fair, or good). To simplify our understanding of how daily gaming duration and academic performance relate within each level of sleep quality, we will simply compare the boxplot for excellent academic performance to the boxplot for failing academic performance in each visualization.

**Very Poor Sleep Quality:** Amongst those in the dataset with very poor sleep quality, we can see that those with excellent work performance have a median of about 6.5 hours of gaming daily with a slightly right skewed distribution, no outliers, and a IQR of about 2 hours between a Q1 of about 5.5 hours gamed and a Q3 of about 7.5 hours gamed daily. On the other hand, those with very poor sleep quality who have a failing academic performance have a higher median of about 8.3 hours of daily gaming, no outliers, a slightly right-skewed distribution for hours gamed daily, and a greater IQR of about 3.3 hours with a Q1 of about 7 and a Q3 of about 10.3 hours of gaming daily. While the spread is greatly different between someone with failing performance and excellent performance for very poor sleep quality, the first visualization shows how the distribution of daily gaming hours is roughly consolidated around similar values regardless of academic performance level, except for good and excellent performance. ***This makes the relationship between academic work performance and daily gaming hours for very poor sleep quality too ambiguous to draw a conclusion yet.***

**Fair Sleep Quality**: For those in the dataset with fair sleep quality, we can see that those with excellent work performance have a median of about 4 hours of daily gaming, a somewhat symmetrical distribution, one high outlier at about 11 hours, and an IQR of 2.8 hours with a Q1 of 2.2 hours and Q3 of 5 hours. On the other hand, those who have failing academic performance with fair sleep quality have a greater spread for daily gaming hours with an IQR of about 4.3 hours from a Q1 of about 5.8 hours and a Q3 of about 10.2 hours. Those with failing performance also have a median of 7.8 hours of daily gaming, no outliers, and a somewhat right skewed distribution. The difference between daily gaming hours for those with excellent and failing performance is greater under fair sleep quality than it was for those with very poor sleep quality. In addition, the second visualization for fair sleep quality shows a more clear,  gradual decrease in the median and middle 50% of daily gaming hours by each performance level compared to the first visualization's distributions. ***This may indicate that a higher sleep quality strengthens a possibly negative relationship between our two variables, where gaming hours decreases as academic performance increases. But, since there are still large spreads and upper and lower outliers in the second visualization for those with average and below average performance level, we can look at a greater level of sleep quality to confirm whether it further strengthens our observed relationship.***

**Good Sleep Quality**: For those in the dataset with good sleep quality, we can see that those with excellent work performance have a lower median of about 3 hours of daily gaming, no outliers, a somewhat right skewed distribution, and an IQR of about 2 hours with a Q1 of about 1.8 hours and a Q3 of about 3.8 hours. On the other hand, those who have a failing performance with a good sleep quality have a median of about 10 hours of daily gaming and a highly left-skewed distribution, but has a smaller range from 4.5 hours to about 10.5 hours of daily gaming compared either of the first two levels of sleep quality. While this reduction in spread under good sleep quality doesn't apply to all other performance levels, we can see a more clear and consistent reduction of the middle 50% of daily gaming hours by performance level, especially when comparing boxplots from good sleep quality to those from very poor sleep quality. ***This growing trend may further confirm that an increase in sleep quality strengthens a negative relationship between academic performance and damily gaming hours.***

**Comparing Sleep Qualities**: To further confirm a decline in daily gaming hours simply based on an increase in sleep quality, we can reference the last dataframe comparing summary statistics of overall daily gaming hours of those from each group in the dataset. When comparing very poor quality sleep to fair and good quality sleep, we can notice that Q1 decreased gradually from 6.6 hours to 2.28 hours of gaming daily and Q3 decreased gradually from 10.00 hours to 5.63 hours of gaming daily. Similarly, the overall mean of daily gaming hours decreased gradually from 8.10 hours to 4.04 hours while the median decreased from 7.80 hours to 3.70 hours daily. These decreases further supports our question about the effects of sleep quality, as it indicates an overall reduction in daily gaming hours as sleep quality increases, even though reduction in daily gaming hours by specific performance level is better modeled by the boxplots.
