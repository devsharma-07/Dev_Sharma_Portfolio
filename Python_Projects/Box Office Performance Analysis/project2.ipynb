# Section 5: ***Data Science***

***Methodology:***

Now, the question we would like to explore is "how well does a model predict worldwide box office collections using opening weekend collections? How often are the predictions overpredictions or underpredictions?" I pivoted the question from section 4 slightly to allow me to accurately use a supervised linear regression model from recent modules to predict worldwide collections based on opening weekend collections. We can first use machine learning through a supervised linear regression model to create predictions of the worldwide box office collections. Then, we can compare the predicted values to the actual worldwide collection values by calculating residuals and percentage errors. We can use the probabilities of there being a negative or positive residual to compare how often overpredictions occur compared to underpredictions, respectively. In addition, to assess how well the model predicts box office collections using opening weekend numbers, we can find the average, median, and standard deviations of the percentage errors.

***Execution:***

First, use machine learning like supervised linear regression, let's add a column to the original dataframe that contains predicted box office collections of each MCU movie based on opening weekend numbers.

```python
from sklearn.linear_model import LinearRegression
ind = ["opening_weekend"]
dep = "worldwide_box_office"
model = LinearRegression()
model.fit(df[ind], df[dep])
df["box_office_predicted"] = model.predict(df[ind])
df
```

Next, we can compare these predicted values to the actual box office numbers by adding new columns that contain residuals (actual-predicted) and percentage error (out of 100%) for each movie.

```python
df["residual"] = df["worldwide_box_office"] - df["box_office_predicted"]
df["percent_error"] = abs((df["box_office_predicted"] - df["worldwide_box_office"])/df["worldwide_box_office"]) * 100
df
```

**Let's compare how often the model overpredicts and underpredicts worldwide box office collections.**

```python
overpredicts_num = len(df[df.residual < 0])
underpredicts_num = len(df[df.residual > 0])
overpredicts_prob = overpredicts_num/len(df)
underpredicts_prob = underpredicts_num/len(df)
print(overpredicts_prob)
print(underpredicts_prob)
```

Conclusion: The model overpredicts the worldwide box office collections for Marvel Universe movies about 55.56% of the time, while it underpredicts them about 44.44% of the time.

**Now, let's see the center and spread for the distribution of percentage errors from using the model to predict worldwide box office collections based on opening weekend numbers.**

```python
print(df["percent_error"].mean())
print(df["percent_error"].median())
print(df["percent_error"].std())
```

Conclusion: The model predicts worldwide box office collections with an average percentage error of 15.17% and a median of 13.12%, indicating that there may be higher percentage errors that skew the average to be greater. This is seen through the standard deviation of 13.59%, which shows a considerably large spread for percent error. This means that the model wasn't consistently effective in predicting the box office collection of Marvel movies, with far worse predictions for some compared to others.
